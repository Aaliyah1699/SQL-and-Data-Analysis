## About Dataset
📂
The **Global Cybersecurity Threats Dataset (2015-2024)** provides extensive data on cyberattacks, malware types, targeted industries, and affected countries. It is designed for threat intelligence analysis, cybersecurity trend forecasting, and machine learning model development to *enhance global digital security*.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Global Cybersecurity Threats Cleaned-2015-2024.csv')

df.info()
```

---

### Which industries experienced the highest and lowest total financial losses due to cyberattacks? 
### What is the most common attack type targeting each industry?

```python
# Set plot styles
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)


# 1. Industries with the Highest and Lowest Total Financial Loss
# Group by 'Target Industry' and sum the 'Financial Loss ($M)'
industry_loss = df.groupby('Target Industry')['Financial Loss ($M)'].sum().sort_values(ascending=False)


# Display the industries with the highest and lowest financial loss
highest_loss_industry = industry_loss.idxmax()
lowest_loss_industry = industry_loss.idxmin()

print(' ')
print(f"Industry with the highest financial loss: {highest_loss_industry} (${industry_loss.max():,.2f}M)")
print(f"Industry with the lowest financial loss: {lowest_loss_industry} (${industry_loss.min():,.2f}M)")
print(' ')

# Visualize the top 10 industries by financial loss
top10_loss = industry_loss.head(10)

# Plot
sns.barplot(x=top10_loss.values, y=top10_loss.index, palette="Reds_r", hue=top10_loss.index, legend=False)
plt.title('Top 10 Industries by Total Financial Loss from Cyberattacks')
plt.xlabel('Total Financial Loss ($ Millions)')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()


# 2. Most Common Attack Type for Each Industry

# Group by 'Target Industry' and 'Attack Type', then count occurrences
industry_attack_counts = df.groupby(['Target Industry', 'Attack Type']).size().reset_index(name='Count')

# For each industry, find the attack type with the maximum count
most_common_attack_per_industry = industry_attack_counts.loc[industry_attack_counts.groupby('Target Industry')['Count'].idxmax()]

# Display results
print("\nMost common attack type for each industry:\n")
print(most_common_attack_per_industry[['Target Industry', 'Attack Type', 'Count']])
print(' ')

plt.figure(figsize=(14, 8))
sns.scatterplot(data=most_common_attack_per_industry, 
                x='Count', 
                y='Target Industry', 
                hue='Attack Type', 
                s=100, 
                palette="viridis")

plt.title('Most Common Attack Type per Industry')
plt.xlabel('Number of Incidents')
plt.ylabel('Industry')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Financial Loss Summary

Analysis of the dataset revealed that the **IT industry** suffered the highest total financial loss from cyberattacks, amounting to **$24.81 billion**.  

In contrast, the **Education sector** experienced the lowest overall financial loss, totaling **$20 billion**.

The industries ranked by total financial loss (from highest to lowest) are as follows:
1. **IT**
2. **Banking**
3. **Government**
4. **Retail**
5. **Healthcare**
6. **Telecommunications**
7. **Education**

This ranking suggests that industries with highly sensitive data and digital infrastructure, such as IT and Banking, are the most financially impacted by cybersecurity threats.  
Meanwhile, industries like Education, despite being targeted, tend to report lower direct financial damages.

### Most Common Attack Types by Industry

The analysis identified the most frequent attack type targeting each industry:

- **Banking**: Phishing (96 incidents)
- **Education**: Distributed Denial of Service (DDoS) (73 incidents)
- **Government**: SQL Injection (75 incidents)
- **Healthcare**: Malware (81 incidents)
- **IT**: DDoS (91 incidents)
- **Retail**: Phishing (89 incidents)
- **Telecommunications**: DDoS (85 incidents)

Key observations:
- **DDoS attacks** are the most common method against industries heavily reliant on continuous service availability (e.g., IT, Education, Telecommunications).
- **Phishing** is the preferred attack vector for industries dealing with financial transactions and personal user information, such as **Banking** and **Retail**.
- **Government entities** are more frequently targeted by **SQL Injection attacks**, likely due to their extensive use of web-based systems.

This trend emphasizes the need for industry-specific cybersecurity strategies, with particular attention to strengthening defenses against the most common forms of attack.

### Which countries have the fastest average resolution times for cybersecurity incidents?

### Which country has incurred the least total financial loss from cybersecurity threats?

### Is there a correlation between resolution time and financial loss across countries?

```python
# --- Fastest Resolution Times by Country ---

# Group by country and calculate the average resolution time
avg_resolution_time = df.groupby('Country')['Resolution Time (Hr)'].mean().sort_values()

# Display the top 10 countries with the fastest (lowest) average resolution times
fastest_resolution_countries = avg_resolution_time.head(10)
print("Top 10 Countries with Fastest Resolution Times:\n")
print(fastest_resolution_countries)

# Visualize the fastest resolution times
plt.figure(figsize=(10,6))
sns.barplot(x=fastest_resolution_countries.values, y=fastest_resolution_countries.index, palette='viridis', hue=fastest_resolution_countries.index, legend=False)
plt.title('Top 10 Countries with Fastest Cyberattack Resolution Times', fontsize=14)
plt.xlabel('Average Resolution Time (Hours)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()


# --- Least Financial Loss by Country ---

# Group by country and calculate total financial loss
total_financial_loss = df.groupby('Country')['Financial Loss ($M)'].sum().sort_values()

# Find the country with the least total financial loss
least_loss_country = total_financial_loss.head(1)
print("\nCountry with the Least Financial Loss:\n")
print(least_loss_country)


# --- Relationship Between Financial Loss and Resolution Time ---

# Merge average resolution time and financial loss into one DataFrame
resolution_vs_loss = pd.merge(
    avg_resolution_time.reset_index(),
    total_financial_loss.reset_index(),
    on='Country'
)

# Scatter plot to check for visual correlation
plt.figure(figsize=(8,6))
sns.scatterplot(data=resolution_vs_loss, x='Resolution Time (Hr)', y='Financial Loss ($M)', hue='Country', legend=False)
plt.title('Resolution Time vs Financial Loss by Country', fontsize=14)
plt.xlabel('Average Resolution Time (Hours)')
plt.ylabel('Total Financial Loss ($M)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate correlation coefficient
correlation = resolution_vs_loss['Resolution Time (Hr)'].corr(resolution_vs_loss['Financial Loss ($M)'])
print(f"\nCorrelation between Resolution Time and Financial Loss: {correlation:.2f}")
```

### Fastest Resolution Times Summary

The top 10 countries with the fastest average resolution times for cybersecurity incidents are as follows:

1. **USA**: 35 hours
2. **Russia**: 35 hours
3. **India**: 35 hours
4. **Germany**: 36 hours
5. **France**: 36 hours
6. **Australia**: 36 hours
7. **UK**: 36 hours
8. **Japan**: 37 hours
9. **China**: 37 hours
10. **Brazil**: 37 hours

The **USA** leads with the fastest average resolution time of **35 hours**, followed closely by **Russia** and **India**. These countries likely have the most advanced cybersecurity infrastructure and response systems, contributing to their faster resolution times.

    
### The Least Financial Loss Summary

The country that suffered the least total financial loss from cybersecurity incidents is **China**, with a total of **$13.71 billion** in losses.

This suggests that, despite having a relatively fast resolution time, **China** incurred less overall financial damage from cyberattacks. This may indicate effective defense measures or lower-targeted industries.


### Correlation Between Resolution Time and Financial Loss

The correlation coefficient between **resolution time** and **financial loss** across countries is **0.17**, which indicates a **weak positive correlation**.

This suggests that, while there is a slight tendency for countries with longer resolution times to experience higher financial losses, the relationship is not strong. Other factors, such as the nature of the attack, the industry targeted, and the country’s cybersecurity infrastructure, likely play a more significant role in determining financial loss.

### Which defense is used most frequently to counter cybersecurity threats?

### Which attack type is most commonly defended against with this defense?

```python
# Step 1: Identify the most commonly used defense method
most_used_defense = df['Defense Used'].value_counts().idxmax()

print(' ')
print(f"Most commonly used defense method: {most_used_defense}")
print(' ')

# Visualize the defense methods usage
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Defense Used', order=df['Defense Used'].value_counts().index, palette='viridis', hue='Defense Used', legend=False)
plt.title('Most Common Defense Methods')
plt.xlabel('Number of Occurrences')
plt.ylabel('Defense Method')
plt.show()

# Step 2: Find out against which attack type this defense is primarily used
# Filter the DataFrame to only rows with the most used defense
filtered_df = df[df['Defense Used'] == most_used_defense]

# Find the most common attack type among these
primary_attack_type = filtered_df['Attack Type'].value_counts().idxmax()

print(' ')
print(f"The primary attack type against which {most_used_defense} is used: {primary_attack_type}")
print(' ')

# Visualize the attack types targeted by this defense
plt.figure(figsize=(10,6))
sns.countplot(data=filtered_df, y='Attack Type', order=filtered_df['Attack Type'].value_counts().index, palette='magma', hue='Attack Type', legend=False)
plt.title(f'Attack Types Targeted by {most_used_defense}')
plt.xlabel('Number of Occurrences')
plt.ylabel('Attack Type')
plt.show()



# --- Most Used Defense ---
# Count the frequency of each defense used across all attacks
most_used_defense = df['Defense Used'].value_counts().head(1)
print(f"Most Used Defense:\n{most_used_defense}")

# --- Most Common Attack Type for Most Used Defense ---
# Filter the data for rows where the most used defense is applied
most_common_attack_for_defense = df[df['Defense Used'] == most_used_defense.index[0]]['Attack Type'].value_counts().head(1)
print(f"\nMost Common Attack Type for {most_used_defense.index[0]} Defense:\n{most_common_attack_for_defense}")
```

### 🛡️ Analysis Summary: Most Commonly Used Defense and Associated Attack Types

**1. Most Commonly Used Defense Methods:**
- The most frequently deployed defense methods were:
  - **Antivirus**
  - **VPN**
  - **Firewall**
  - **AI-Based Detection**

These defenses were used at similar rates, with **Antivirus** being slightly more prominent overall.

---

**2. Most Common Attack Types Defended by Antivirus:**
- When focusing specifically on the **Antivirus** defense method, the primary cyberattack types it was used against were:
  - **SQL Injection**
  - **Phishing**
  - **Malware** (used about as often as Phishing)
  - **Ransomware**
  - **DDoS Attacks** (used about as often as Ransomware)
  - **Man-in-the-Middle (MITM) Attacks**

These findings highlight that **Antivirus** solutions are broadly applied across both traditional malware-based threats and sophisticated network attacks.

---

> **Insight:**  
> Organizations appear to rely on a combination of traditional and AI-based defenses to protect against a wide range of attacks, suggesting a strategy that blends legacy security solutions with emerging technologies.

### Which attack type is most commonly associated with each attack source?

### Which attack type generates the highest financial loss for the attack sources?

```python
# --- Most Common Attack Type for Each Attack Source ---
# Group by attack source and attack type, then count occurrences for each combination
attack_source_attack_type = df.groupby(['Attack Source', 'Attack Type']).size().reset_index(name='Count')

# Find the most common attack type for each attack source
most_common_attack_per_source = attack_source_attack_type.loc[attack_source_attack_type.groupby('Attack Source')['Count'].idxmax()]
print(f"Most Common Attack Type for Each Attack Source:\n{most_common_attack_per_source}")

# --- Attack Type Generating the Highest Financial Loss for Attack Sources ---
# Group by attack type and calculate total financial loss per attack type
attack_type_loss = df.groupby('Attack Type')['Financial Loss ($M)'].sum().reset_index()

# Find the attack type generating the highest financial loss
highest_loss_attack_type = attack_type_loss.loc[attack_type_loss['Financial Loss ($M)'].idxmax()]
print(f"\nAttack Type Generating the Highest Financial Loss:\n{highest_loss_attack_type}")
```

### Most Common Attack Type for Each Attack Source

The most common attack type associated with each attack source is as follows:

- **Hacker Group**: Phishing (125 occurrences)
- **Insider**: DDoS (147 occurrences)
- **Nation-state**: Phishing (141 occurrences)
- **Unknown**: SQL Injection (141 occurrences)

This indicates that **Phishing** is the most common attack type for both **Hacker Groups** and **Nation-states**, while **DDoS** is the primary attack type for **Insiders**. The **Unknown** attack source is most commonly associated with **SQL Injection**, suggesting that unidentified sources tend to leverage vulnerabilities in databases.

### Attack Type Generating the Highest Financial Loss

The attack type that generates the highest financial loss is **DDoS**, with a total financial loss of **$27.63 billion**.

This suggests that **DDoS** attacks are not only frequent but also particularly costly, likely due to their widespread impact on business operations, infrastructure, and service disruptions, leading to significant financial damage.

The **Education** sector, while targeted frequently by **Nation-states**, faces relatively lower financial loss in comparison.

### In which year were the most and least users affected by cyberattacks?

### In which year(s) were the highest and lowest financial losses recorded?

```python

# --- Year with Most and Least Affected Users ---
# Group by year and calculate the total number of affected users per year
year_affected_users = df.groupby('Year')['Number of Affected Users'].sum().reset_index()

# Identify the year with the most and least affected users
most_affected_year = year_affected_users.loc[year_affected_users['Number of Affected Users'].idxmax()]
least_affected_year = year_affected_users.loc[year_affected_users['Number of Affected Users'].idxmin()]
print(f"Year with the Most Affected Users: {most_affected_year}")
print(f"Year with the Least Affected Users: {least_affected_year}")

# --- Year with Highest and Lowest Financial Loss ---
# Group by year and calculate the total financial loss per year
year_financial_loss = df.groupby('Year')['Financial Loss ($M)'].sum().reset_index()

# Identify the year with the highest and lowest financial loss
highest_loss_year = year_financial_loss.loc[year_financial_loss['Financial Loss ($M)'].idxmax()]
lowest_loss_year = year_financial_loss.loc[year_financial_loss['Financial Loss ($M)'].idxmin()]
print(f"\nYear with the Highest Financial Loss: {highest_loss_year}")
print(f"Year with the Lowest Financial Loss: {lowest_loss_year}")

# --- Visualization: Affected Users per Year ---
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Number of Affected Users', data=year_affected_users, palette='viridis', hue='Number of Affected Users', legend=False)
plt.title('Total Number of Affected Users per Year')
plt.xlabel('Year')
plt.ylabel('Total Affected Users')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- Visualization: Financial Loss per Year ---
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Financial Loss ($M)', data=year_financial_loss, palette='magma', hue='Financial Loss ($M)', legend=False)
plt.title('Total Financial Loss per Year')
plt.xlabel('Year')
plt.ylabel('Total Financial Loss ($M)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 📊 Analysis Summary: Users Affected and Financial Loss by Year

**1. Most and Least Affected Users:**
- **Most Affected Year:**  
  - **Year:** 2022  
  - **Number of Affected Users:** 163,263,160

- **Least Affected Year:**  
  - **Year:** 2019  
  - **Number of Affected Users:** 130,212,928

This shows a significant increase in cyberattack impact on users from 2019 to 2022, highlighting an upward trend in large-scale cyber incidents.

**2. Highest and Lowest Financial Loss:**

### Which security vulnerability type has resulted in the highest financial loss across industries?

### What is the average resolution time (in hours) for each target industry?

Other major contributing vulnerabilities include:

**Possible Preventative Solutions:**
- **Zero-Day Vulnerabilities:**  
  - Implement real-time threat intelligence systems.
  - Regularly update and patch systems, even proactively when exploits are disclosed.

- **Social Engineering Attacks:**  
  - Conduct continuous employee cybersecurity awareness and phishing simulation training.
  - Establish strict email verification protocols.

- **Unpatched Software:**  
  - Enforce automated patch management processes.
  - Maintain an inventory of all software assets, with patch statuses tracked.

- **Weak Passwords:**  
  - Mandate multifactor authentication (MFA) across all systems.
  - Enforce strong password policies with regular password audits.

### Is there a relationship between resolution time and financial loss?

### Is there a relationship between resolution time and the number of affected users?

```python
# Step 1: Scatter plot to check relationship between Resolution Time and Financial Loss
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Resolution Time (Hr)', y='Financial Loss ($M)', hue='Attack Type', alpha=0.7)
plt.title('Resolution Time vs Financial Loss')
plt.xlabel('Resolution Time (Hours)')
plt.ylabel('Financial Loss ($M)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

print(' ')
# Step 2: Calculate the correlation between Resolution Time and Financial Loss
corr_loss = df['Resolution Time (Hr)'].corr(df['Financial Loss ($M)'])
print(f"Correlation between Resolution Time and Financial Loss: {corr_loss:.3f}")
print(' ')

# Step 3: Scatter plot to check relationship between Resolution Time and Number of Affected Users
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Resolution Time (Hr)', y='Number of Affected Users', hue='Attack Type', alpha=0.7)
plt.title('Resolution Time vs Number of Affected Users')
plt.xlabel('Resolution Time (Hours)')
plt.ylabel('Number of Affected Users')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

print(' ')
# Step 4: Calculate the correlation between Resolution Time and Number of Affected Users
corr_users = df['Resolution Time (Hr)'].corr(df['Number of Affected Users'])
print(f"Correlation between Resolution Time and Number of Affected Users: {corr_users:.3f}")
print(' ')
```

### 🔎 Analysis Summary: Impact of Resolution Time on Financial Loss and Affected Users

**Key Findings:**

1. **Resolution Time vs Financial Loss**
   - **Correlation Coefficient:** -0.013
   - **Interpretation:**  
     There is virtually no correlation between how quickly an incident is resolved and the amount of financial loss incurred.  
     A negative value very close to 0 suggests that faster or slower resolution times do not consistently reduce or increase financial loss.

2. **Resolution Time vs Number of Affected Users**
   - **Correlation Coefficient:** 0.006
   - **Interpretation:**  
     Similarly, there is no meaningful relationship between resolution time and the number of users affected.  
     The extremely low positive value indicates that faster resolution does not significantly limit the number of users impacted.

**Visual Observations:**
- Scatter plots showed a **highly scattered distribution** with no obvious trends or patterns.
- Points were spread randomly, confirming the correlation results statistically.

---

> **Overall Interpretation:**  
> Quick resolution times, while critical operationally, do not by themselves guarantee lower financial loss or fewer affected users based on this dataset. Other factors such as **attack type, industry preparedness,** and **vulnerability exploited** may play a stronger role in determining the impact of cybersecurity incidents.
