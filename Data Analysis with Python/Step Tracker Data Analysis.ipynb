```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('CSV Files/steps_tracker_dataset.csv')
```

```python
 df.columns
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df['steps'] = df['steps'].astype(int)
df['calories_burned'] = df['calories_burned'].astype(float)
df['sleep_hours'] = df['sleep_hours'].astype(float)
df['active_minutes'] = df['active_minutes'].astype(int)
df['water_intake_liters'] = df['water_intake_liters'].astype(float)
df['distance_km'] = df['distance_km'].astype(float)
```

```python
# Group by 'mood' and calculate average of 'steps', 'calories_burned', and 'sleep_hours'
mood_group = df.groupby('mood').agg({
    'steps': 'mean',
    'calories_burned': 'mean',
    'sleep_hours': 'mean'
}).reset_index()

# Violin plot for mood-based grouping
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot 1: Steps by mood
sns.violinplot(ax=axes[0], x='mood', y='steps', data=df, hue='mood', inner="quart", palette="Paired", legend=False)
axes[0].set_title('Violin plot of Steps by Mood')

# Plot 2: Calories by mood
sns.violinplot(ax=axes[1], x='mood', y='calories_burned', data=df, hue='mood', inner="quart", palette="tab10", legend=False)
axes[1].set_title('Violin plot of Calories Burned by Mood')

# Plot 3: Sleep Hours by mood
sns.violinplot(ax=axes[2], x='mood', y='sleep_hours', data=df, hue='mood', inner="quart", palette="Dark2", legend=False)
axes[2].set_title('Violin plot of Sleep Hours by Mood')

plt.tight_layout()
plt.show()
```

```python
# Categorize days based on sleep hours
bins = [0, 5, 7, 24]  # Define the sleep hour ranges
labels = ['Low Sleep', 'Moderate Sleep', 'High Sleep']  # Labels for the categories
df['sleep_category'] = pd.cut(df['sleep_hours'], bins=bins, labels=labels, right=False)

# Scatter plot to analyze how sleep hours affect steps and mood
plt.figure(figsize=(10, 6))

sns.scatterplot(x='sleep_hours', y='steps', hue='sleep_category', style='mood', data=df, palette='Set1')

plt.title('Scatter Plot of Steps vs Sleep Hours by Mood')
plt.xlabel('Sleep Hours')
plt.ylabel('Steps')
plt.legend(title='Sleep Category')
plt.tight_layout()
plt.show()
```

```python
# Create a figure with two subplots placed next to each other
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Scatter Plot 1: Steps vs Distance
sns.scatterplot(ax=axes[0], x='steps', y='distance_km', data=df, color='skyblue')
axes[0].set_title('Steps vs Distance')
axes[0].set_xlabel('Steps')
axes[0].set_ylabel('Distance (km)')

# Scatter Plot 2: Steps vs Calories Burned
sns.scatterplot(ax=axes[1], x='steps', y='calories_burned', data=df, color='lightcoral')
axes[1].set_title('Steps vs Calories Burned')
axes[1].set_xlabel('Steps')
axes[1].set_ylabel('Calories Burned')

# Adjust the layout for better spacing between the plots
plt.tight_layout()
plt.show()
```

```python
# Create a figure with two subplots placed next to each other
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Violin Plot 1: Sleep Hours vs Mood
sns.violinplot(ax=axes[0], x='mood', y='sleep_hours', hue='mood', data=df, inner="quart", palette="Set1")
axes[0].set_title('Sleep Hours vs Mood')
axes[0].set_xlabel('Mood')
axes[0].set_ylabel('Sleep Hours')

# Box Plot 2: Water Intake vs Mood
sns.boxplot(ax=axes[1], x='mood', y='water_intake_liters', hue='mood', data=df, palette="tab20")
axes[1].set_title('Water Intake vs Mood')
axes[1].set_xlabel('Mood')
axes[1].set_ylabel('Water Intake (liters)')

# Adjust the layout for better spacing between the plots
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Active Minutes vs Calories Burned
plt.figure(figsize=(8, 6))

sns.scatterplot(x='active_minutes', y='calories_burned', data=df, color='purple')

plt.title('Scatter Plot of Active Minutes vs Calories Burned')
plt.xlabel('Active Minutes')
plt.ylabel('Calories Burned')
plt.tight_layout()
plt.show()
```

```python
# 1. Calculate Mean (sleep_hours, active_minutes)
mean_values = df[['sleep_hours', 'active_minutes']].mean()

# 2. Calculate Median (water_intake_liters, distance_km)
median_values = df[['water_intake_liters', 'distance_km']].median()

# 3. Calculate Mode (mood, date (month))
mode_values_mood = df['mood'].mode()[0]
mode_values_month = df['date'].dt.month.mode()[0]

# Create a figure with three subplots placed next to each other
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 1. Bar Chart for Mean (sleep_hours, active_minutes)
mean_values.plot(kind='bar', ax=axes[0], color=['skyblue', 'lightgreen'])
axes[0].set_title('Mean of Sleep Hours and Active Minutes')
axes[0].set_xlabel('Variables')
axes[0].set_ylabel('Mean Value')
axes[0].set_xticklabels(['Sleep Hours', 'Active Minutes'], rotation=0)

# 2. Box Plot for Median (water_intake, distance_km)
sns.boxplot(ax=axes[1], data=[df['water_intake_liters'], df['distance_km']], palette="Set2")
axes[1].set_title('Median of Water Intake and Distance')
axes[1].set_ylabel('Values')

# 3. Pie Chart for Mode (mood, date (month))
# Mood Mode
mood_counts = df['mood'].value_counts()
axes[2].pie(mood_counts, labels=mood_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Dark2', len(mood_counts)))
axes[2].set_title(f'Mode of Moods and Months: {mode_values_month}')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# 1. Calculate Min (steps, water_intake)
min_values = [df['steps'].min(), df['water_intake_liters'].min()]

# 2. Calculate Max (calories_burned, distance_km)
max_values = [df['calories_burned'].max(), df['distance_km'].max()]

# Create a figure with two subplots placed next to each other
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 1. Point Plot for Min (steps, water_intake)
axes[0].scatter(['Steps', 'Water Intake (liters)'], min_values, color='navy', s=100, zorder=5)
axes[0].set_title('Min of Steps and Water Intake')
axes[0].set_ylabel('Min Value')
axes[0].grid(True, linestyle='--', alpha=0.7)

# 2. Horizontal Bar Chart for Max (calories_burned, distance_km)
axes[1].barh(['Calories Burned', 'Distance (km)'], max_values, color=['slateblue', 'royalblue'])
axes[1].set_title('Max of Calories Burned and Distance')
axes[1].set_xlabel('Max Value')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
