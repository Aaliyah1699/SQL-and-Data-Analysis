```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
sales = pd.read_csv('CSV Files/superstore_sales.csv')
# Ensure date columns are parsed correctly (using dayfirst for DD/MM/YYYY format)
sales['Order Date'] = pd.to_datetime(sales['Order Date'], dayfirst=True, errors='coerce')
sales['Ship Date'] = pd.to_datetime(sales['Ship Date'], dayfirst=True, errors='coerce')
```

```python
# Sales Performance Analysis

# Revenue Trend Analysis: Aggregate sales by month
# Create a new column for year-month
sales['Year_Month'] = sales['Order Date'].dt.to_period('M')
monthly_sales = sales.groupby('Year_Month')['Sales'].sum().reset_index()
monthly_sales['Year_Month'] = monthly_sales['Year_Month'].astype(str)

# Plot the total sales over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Year_Month', y='Sales', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()

# Profitability Analysis: Total profit by Category and Sub-Category
# Aggregate profit by Category and Sub-Category
profit_by_category = sales.groupby(['Category', 'Sub-Category'])['Profit'].sum().reset_index()

# Plot a bar chart for profit by sub-category within each category
plt.figure(figsize=(12, 6))
sns.barplot(data=profit_by_category, x='Sub-Category', y='Profit', hue='Category')
plt.title('Profit by Category and Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Customer Segmentation Analysis


# Sales by Segment: Determine how sales distribute across customer segments
sales_by_segment = sales.groupby('Segment')['Sales'].sum().reset_index()

# Plot a bar chart of sales distribution by segment
plt.figure(figsize=(8, 6))
sns.barplot(data=sales_by_segment, x='Segment', y='Sales', hue='Segment', palette='viridis', legend=False)
plt.title('Sales Distribution by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Customer Lifetime Value (CLV) Analysis: Total sales and profit per customer
clv = sales.groupby('Customer ID').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Scatter plot: Total Sales vs. Total Profit per Customer
plt.figure(figsize=(10, 6))
sns.scatterplot(data=clv, x='Sales', y='Profit', alpha=0.6)
plt.title('Customer Lifetime Value: Sales vs. Profit')
plt.xlabel('Total Sales per Customer')
plt.ylabel('Total Profit per Customer')
plt.tight_layout()
plt.show()
```

```python
# Product Performance Analysis

# Top-Selling Products: Identify top 10 products by sales
product_sales = sales.groupby('Product Name')['Sales'].sum().reset_index()
top_products = product_sales.sort_values(by='Sales', ascending=False).head(10)

# Horizontal bar chart for top-selling products
plt.figure(figsize=(12, 6))
sns.barplot(data=top_products, x='Sales', y='Product Name', hue='Product Name', palette='coolwarm', legend=False)
plt.title('Top 10 Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()

# High-Profit vs. Low-Profit Products: Scatter plot comparing sales vs. profit for products
product_performance = sales.groupby('Product Name').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

plt.figure(figsize=(12, 6))
sns.scatterplot(data=product_performance, x='Sales', y='Profit', alpha=0.6)
plt.title('Product Performance: Sales vs. Profit')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.tight_layout()
plt.show()
```

```python
# Discount Impact Analysis

# Discount vs. Profitability Analysis
# Compute profit margin as Profit divided by Sales
sales['Profit_Margin'] = np.where(sales['Sales'] != 0, sales['Profit'] / sales['Sales'], np.nan)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=sales, x='Discount', y='Profit_Margin', alpha=0.6)
plt.title('Discount vs. Profit Margin')
plt.xlabel('Discount')
plt.ylabel('Profit Margin')
plt.tight_layout()
plt.show()

# Average Discount by Category and Sub-Category
plt.figure(figsize=(12, 6))
sns.boxplot(data=sales, x='Category', y='Discount', hue='Sub-Category')
plt.title('Distribution of Discounts by Category and Sub-Category')
plt.xlabel('Category')
plt.ylabel('Discount')
plt.legend(title='Sub-Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Shipping Performance Analysis

# Shipping Delays Analysis: Calculate processing time in days (Ship Date - Order Date)
sales['Processing_Time'] = (sales['Ship Date'] - sales['Order Date']).dt.days

plt.figure(figsize=(10, 6))
sns.histplot(sales['Processing_Time'].dropna(), bins=30, kde=False)
plt.title('Distribution of Order Processing Times')
plt.xlabel('Processing Time (Days)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Sales by Ship Mode: Analyze how sales are distributed by shipping mode
sales_by_shipmode = sales.groupby('Ship Mode')['Sales'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=sales_by_shipmode, x='Ship Mode', y='Sales', hue='Ship Mode', palette='muted', legend=False)
plt.title('Total Sales by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Regional Sales Performance

# Sales by Region: Create a choropleth map for US states
state_to_abbrev = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR',
    'California': 'CA', 'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE',
    'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID',
    'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS',
    'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
    'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
    'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY',
    'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
    'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT',
    'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV',
    'Wisconsin': 'WI', 'Wyoming': 'WY'
}

# Create a new column for state abbreviations. 
sales['State_Abbrev'] = sales['State'].map(state_to_abbrev)

# Aggregate sales by state abbreviation
sales_by_state = sales.groupby('State_Abbrev')['Sales'].sum().reset_index()

# create a choropleth map of US states
fig = px.choropleth(
    sales_by_state,
    locations='State_Abbrev',  # Column with state abbreviations
    locationmode="USA-states",
    color='Sales',
    scope="usa",
    title='Total Sales by US State',
    color_continuous_scale='Blues'
)
fig.show()

# Profitability by State: Visualize profit per state using a bar chart
profit_by_state = sales.groupby('State')['Profit'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=profit_by_state, x='State', y='Profit', hue='State', palette='coolwarm', legend=False)
plt.title('Total Profit by State')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Demand Forecasting (Time Series Analysis)

# Time Series Analysis: Calculate monthly sales and its rolling average
# Create a 'Year_Month'
sales['Year_Month'] = sales['Order Date'].dt.to_period('M')
monthly_sales = sales.groupby('Year_Month')['Sales'].sum().reset_index()
# Convert the Period to string for plotting
monthly_sales['Year_Month'] = monthly_sales['Year_Month'].astype(str)

# Calculate a 3-month rolling average of sales
monthly_sales['Rolling_Avg'] = monthly_sales['Sales'].rolling(window=3).mean()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Year_Month'], monthly_sales['Sales'], marker='o', label='Total Sales')
plt.plot(monthly_sales['Year_Month'], monthly_sales['Rolling_Avg'], marker='o', label='3-Month Rolling Avg', linestyle='--')
plt.title('Monthly Sales and 3-Month Rolling Average')
plt.xlabel('Year-Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
